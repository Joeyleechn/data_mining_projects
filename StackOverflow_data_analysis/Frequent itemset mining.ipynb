```python
import pandas as pd
from apyori import apriori
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df = pd.read_csv('posts_questions_2020.csv')
```

### All programming languages 

```python
question_all = df
question_all['tag_list'] = question_all['tags'].apply(lambda x: x.split('|'))
question_all['list_len'] = question_all['tag_list'].apply(lambda x: len(x))
question_all = question_all[question_all['list_len'] > 1]

records_all = []
for i in range(0, len(question_all.index)):
    records_all.append(question_all['tag_list'][question_all.index[i]])
```

```python
association_rules_all = apriori(records_all, min_support=0.05, min_confidence=0, min_lift=0, min_length=2)
association_results_all = list(association_rules_all)
```

### python

```python
question_python = df[df['tags'].str.contains('python')]
question_python['tag_list'] = question_python['tags'].apply(lambda x: x.split('|'))
question_python['list_len'] = question_python['tag_list'].apply(lambda x: len(x))
question_python = question_python[question_python['list_len'] > 1]

records_python = []
for i in range(0, len(question_python.index)):
    records_python.append(question_python['tag_list'][question_python.index[i]])
```

```python
association_rules_python = apriori(records_python, min_support=0.008, min_confidence=0, min_lift=0, min_length=2)
association_results_python = list(association_rules_python)
```
